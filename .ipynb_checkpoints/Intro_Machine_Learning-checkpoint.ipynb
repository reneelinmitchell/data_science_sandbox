## Building my first ML model since 2020

```python
import pandas as pd
```

```python
melbourne_file_path = 'data/sandbox_data/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
melbourne_data.describe()
```

```python
melbourne_data.columns
```

```python
## Dropna will drop missing values (na || N/A || not available)

melbourne_data = melbourne_data.dropna(axis = 0)
```

```python
melbourne_data.describe()
```

```python
melbourne_data.head()
```

```python
print(melbourne_data.Price)
```

```python
print(melbourne_data.Price[:3])
```

```python
## defining the features that are relevant to the model

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

## by convention, this data is called X

X = melbourne_data[melbourne_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

# apparently this is called defining a model
melbourne_model = DecisionTreeRegressor(random_state = 1)
```

```python
# apparently this is fitting the model
y = melbourne_data.Price

melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

## Model validation

Model validation measures the quality of a model. Will the model's predictions be close to what actually happens?

One such metric for summarizing model quality is Mean Absolute Error (MAE), `mean_absolute_error`, to take the absolute value of error - converting each error to a positive number and then take the average of these errors:

$error = actual - predicted$

```python
from sklearn.metrics import mean_absolute_error

# make the prediction
predicted_home_prices = melbourne_model.predict(X)

# find the MAE of the prediction
mean_absolute_error(y, predicted_home_prices)
```

```python
## however, not good to just use one sample of data for building the model and evaluating
### recommended to exclude some data from the model-building process and then use those to test the model's accuracy on data it hasn't seen before
### this is called validation data

## below we use train_test_split
from sklearn.model_selection import train_test_split

# splitting data into training and validation data
# the split is based on a random number generator
# supplying a numeric value to the random_state argument guarantees we get the same split every time we run this script

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

print("train_X:\n")
print(train_X)

print("val_X:\n")
print(val_X)

print("train_y:\n")
print(train_y)

print("val_y:\n")
print(val_y)
```

```python
## Define the model
melbourne_model = DecisionTreeRegressor()
melbourne_model
```

```python
## Fit the model using the training data
melbourne_model.fit(train_X, train_y)
```

```python
## Get predicted priess on validation data
val_predictions = melbourne_model.predict(val_X)

print(mean_absolute_error(val_y, val_predictions))
```

**Overfitting**: Where a model matches the training data almost perfectly but does poorly in validation and other new data (low error in training but high error in validation) - captures patterns that won't recur in the future

**Underfitting**: When a model fails to capture important distinctions and patterns in the data (high error in both training and validation)

*Both lead to less accurate predictions*
